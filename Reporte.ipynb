# **Analisís inicial de la calidad de los datos**

## **Diccionario de Datos**

Este diccionario de datos nos permitira entender, en una primera instancia, de que se trata cada uno de los campos de las distintas tablas.

### **Tabla Venta**

Esta tabla contiene toda la informacion referida a las ventas y contiene los siguientes campos:

• **IdVenta:** Se trata de un numero de identificacion de la venta (Transaccion), este numero es unico por cada venta.

• **Fecha:** Fecha en la cual se realizo la venta.

• **Fecha_Entrega:** Fecha en la cual fue entregado el producto al cliente.

• **IdCanal:** Numero que identifica el canal por el cual se concreto la venta (Online, Telefonico, etc)

• **IdCliente:** Numero por el cual se identifica el cliente.

• **IdSucursal:** Numero por el cual se identifica a la sucursal que genero la venta.

• **IdEmpleado:** Numero por el cual se identifica al empleado que concreto la venta.

• **IdProducto:** Numero por el cual se identifica a un producto.

• **Precio:** Precio por unidad del producto.

• **Cantidad:** Cantidad de productos involucrados en esa transaccion.

### **Tabla Compra**

Esta tabla contiene toda la informacion  correspondientes a las compras a proveedores.

• **IdCompra:** Este es el numero de identificacion unico de cada compra a proveedor. Este numero es unico.

• **Fecha:** Fecha en la que se realizo la compra del producto al proovedor.

• **Fecha_Año:** Año en el que se realizo la compra al proveedor.

• **Fecha_Mes:** Mes en el que se realizo la compra al proveedor.

• **Fecha_Periodo:** Periodo en el que se realizo la compra al proveedor.

• **IdProducto:** Numero por el cual se identifica a un producto.

• **Cantidad:** Cantidad de productos involucrados en esa transaccion.

• **Precio:** Precio por unidad del producto.

• **IdProveedor:** Numero de identificacion particular de cada proveedor.

### **Tabla Gastos**

Esta tabla contiene toda la informacion correspondientes a los gastos. Esto involucra montos, fechas, tipos y surcusal involucrada.

• **IdGasto:** Este numero identifica a cada gasto realizado.

• **IdSucursal:** Este numero sirve para identificar a la sucursal en la que se realizo el gasto.

• **IdTipoGasto:** Este numero identifica el tipo de gasto en cuestion.

• **Fecha:** Fecha en la que se realizo el gasto.

• **Monto:** Nos indica la cantidad de dinero gastada.

### **Tabla Cliente**

Esta tabla contiene toda la informacion correspondiente a los clientes.

• **ID:** Este numero sirve para identificar a cada cliente en particular.

• **Provincia:** Nos indica la provincia del cliente.

• **Nombre_y_Apellido** Nos indica el Nombre y Apellido del cliente registrado.

• **Domicillio:** Nos indica el domicillio del cliente.

• **Telefono:** Nos indica el numero de telefono del cliente.

• **Edad:** Nos indica la edad del cliente.

• **Localidad:** Este campo nos indica la localidad del cliente.

• **X:** Nos indica la longitud (Coordenada) en la cual se encuentra la localidad del cliente.

• **Y:** Nos indica la latitud (Coordenada) en la cual se encuentra la localidad del cliente.

### **Tabla Localidad**

Esta tabla contiene toda la informacion referente a las distintas localidades registradas.

• **Categoria:** Nos indica si se trata de una localidad simple o compuesta.

• **centroide_lat:** Nos indica la latitud (Coordenada) en la cual se encuentra la localidad.

• **centroide_lon:** Nos indica la longitud (Coordenada) en la cual se encuentra la localidad.

• **departamento_id:** Este numero nos indica el numero de departamento (zona). Cada departamento tiene su propio numero de ID.

• **departamento_nombre:** Este numero nos indica el nombre del departamento.

• **fuente:** Esto nos indica la fuente de la informacion de dicha entrada.

• **id:** Numero unico compuesto por los los ids de departamento, localidad y municipio.

• **localidad_censal_id:** Numero de identificacion de la localidada censal.

• **localidad_censal_nombre:** Nombre de la localidad censal.

• **municipio_id:** Numero de identificacion del municipio al que corresponde la localidad.

• **municipio_nombre:** Nombre del municipio al que corresponde la localidad.

• **nombre:** Nombre de la localidad.

• **id_provincia:** Numero de identificacion de cada provincia.

• **provincia:** Nombre de la provincia correspondiente a la localidad.

### **Tabla Proveedor**

Esta tabla contiene toda la informacion registrada de los distintos proveedores.

• **IDProveedor:** Numero de identificacion unico de cada proveedor.

• **Nombre:** Nombre del proveedor.

• **Address:** Domicillio del proveedor.

• **City:** Localidad en la que se encuentra el proveedor.

• **State:** Provincia en la cual se encuentra el proveedor.

• **Country:** Pais del proveedor.

• **departamen:** Departamento en el que se encuentra el proveedor.

### **Tabla Sucursal**

Esta tabla contiene toda la informacion registrada sobre las distintas sucursales.

• **ID:** Numero unico de identificacion de la sucursal.

• **Sucursal:** Nombre de la sucursal.

• **Direccion:** Direccion de la sucursal.

• **Localidad:** Localidad de la que forma parte la sucursal.

• **Provincia:** Provincia en la cual se encuentra la sucursal.

• **Latitud:** Latitud (Coordenada) de la sucursal.

• **Longitud:** Longitud (Coordenada) de la sucursal.

### **Tabla Canal**

Esta tabla contiene toda la informacion sirve para categorizar los canales de venta.

• **CODIGO:** Este es el numero para identificar un tipo de venta.

• **DESCRIPCION:** Este campo contiene el tipo de canal de venta.

### **Tabla TipoDeGasto**

• **IdTipoGasto:** Numero de asociado a cada tipo de gasto.

• **Descripcion:** Este campo contiene una breve descripcion del tipo de gasto

• **Monto_Aproximado:** Este campo contiene un monto estimado para cada tipo de gasto.

### **Al observar el diccionario de datos lo primero que salta a la vista es que el nombre de los distintos campos de las tablas no tiene sigue una regla clara en cada una de ellas. Esto puede generar dificultades a la hora de realizar consultas.**
